```python
import re
import pandas as pd

# Đọc nội dung file txt
with open("dulieumbti.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()

ind = 0 
s = str.format(raw_text)
arr = []
for i in range(1, len(s)):
    if s[i].isupper():
        row = s[ind: i]
        arr.append(row.strip())
        ind = i

rs = []
for i in range(0, len(arr), 3):
    if i % 3 == 0 and i < len(arr)-3:
        rs.append([arr[i], arr[i+1], arr[i+2]])

# Xuất ra DataFrame
df = pd.DataFrame(rs, columns=["question", "ans1", "ans2"])
print(df)
# # Lưu file
df.to_excel("answers.xlsx", index=False)
# df.to_csv("answers.csv", index=False, encoding="utf-8-sig")

# print("✅ Đã tạo answers.xlsx và answers.csv thành công!")
```
